```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
tesla=yf.Ticker("TSLA")
```

```python
tesla_data=tesla.history(period="max")
tesla_data.head()
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"

response = requests.get(url)

html_data = response.text
print(html_data[:500])

```

```python
soup = BeautifulSoup(html_data, 'html.parser')
print(soup.title)

# Extract links (optional)
for link in soup.find_all('a'):
    print(link.get('href'))

# Initialize DataFrame
Revenue_data_tesla = pd.DataFrame(columns=["Date", "Revenue"])

# Find the table and extract data
table = soup.find('table')
for row in table.find("tbody").find_all('tr'):
    col = row.find_all("td")
    date = col[0].text.strip()
    revenue = col[1].text.strip()
    Revenue_data_tesla = pd.concat([Revenue_data_tesla, pd.DataFrame({"Date": [date], "Revenue": [revenue]})], ignore_index=True)

# Clean the revenue data
Revenue_data_tesla["Revenue"] = Revenue_data_tesla["Revenue"].str.replace(',', '').str.replace('$', '')
Revenue_data_tesla.dropna(inplace=True)
Revenue_data_tesla = Revenue_data_tesla[Revenue_data_tesla["Revenue"] != ""]

Revenue_data_tesla.tail()
```

```python
GameStop=yf.Ticker("GME")
gme_data=GameStop.history(period="max")
gme_data.head()
```

```python
Url=" https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
response=requests.get(Url)
html_data2=response.text
html_data2[:500]

soup2 = BeautifulSoup(html_data2, 'html.parser')

print(soup2.title)
Revenue_data_gme = pd.DataFrame(columns=["Date", "Revenue"])

table2=soup.find('table')

for row in table2.find("tbody").find_all('tr'):
    col = row.find_all("td")
    date = col[0].text
    revenue = col[1].text
    # Agregar la fila al dataframe
    Revenue_data_gme = pd.concat([Revenue_data_gme,pd.DataFrame({"Date":[date], "Revenue":[revenue]})], ignore_index=True)


Revenue_data_gme["Revenue"] = Revenue_data_gme["Revenue"].astype(str).str.replace('[\$,]', '', regex=True)
Revenue_data_gme.dropna(inplace=True)
Revenue_data_gme = Revenue_data_gme[Revenue_data_gme["Revenue"] != ""]

Revenue_data_gme.tail()
```

```python
make_graph(gme_data, Revenue_data_gme, 'GME')
```

```python
def make_graph(stock_data, revenue_data, stock):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Historical Share Price", "Historical Revenue"), vertical_spacing=.3)
    stock_data_specific = stock_data[stock_data.index <= '2021-06-14']
    revenue_data_specific = revenue_data[revenue_data["Date"] <= '2021-04-30']
    fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data_specific.index, infer_datetime_format=True), y=stock_data_specific["Close"].astype("float"), name="Share Price"), row=1, col=1)
    fig.add_trace(go.Scatter(x=pd.to_datetime(revenue_data_specific["Date"], infer_datetime_format=True), y=revenue_data_specific["Revenue"].astype("float"), name="Revenue"), row=2, col=1)
    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)
    fig.update_yaxes(title_text="Price ($US)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=1)
    fig.update_layout(showlegend=False, height=900, title=stock, xaxis_rangeslider_visible=True)
    fig.show()
    fig_html = fig.to_html()
    display(HTML(fig_html))

# Call the function
make_graph(tesla_data, Revenue_data_tesla, 'Tesla')
make_graph(gme_data, Revenue_data_gme, 'GME')
```
